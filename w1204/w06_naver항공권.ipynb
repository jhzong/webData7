```python
# 화면스크롤 내림
prev_height=browser.execute_script("return document.body.scrollHeight")
prev_height
```

```python
# 스크롤 아래로 내리기
while True:
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(3)# 추가데이터 로딩시간
    curr_height=browser.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height:
        break
    # 같지 않으면
    prev_height=curr_height
    print('높이변화 :',curr_height)

print('스크롤 완료 :',curr_height)
```

```python
title=['항공사명','출발시간','도착시간','가격']
# csv파일로 저장
f= open('flightData.csv',"w",encoding="utf-8-sig",newline="")
writer=csv.writer(f)

writer.writerow(title)
```

```python
# 항공편 데이터 전체 출력
for flight in flights:
    f_name=flight.find('b',{'class':'airline_name__0Tw5w'}).text.strip()
    d_time=flight.find_all('b',{'class':'route_time__xWu7a'})[0].text.strip()
    a_time=flight.find_all('b',{'class':'route_time__xWu7a'})[1].text.strip()
    price=int(flight.find('i',{'class':'domestic_num__ShOub'}).text.strip().replace(',',''))
    print(f'{f_name}\t{d_time}\t{a_time}\t{price}')
    print('-'*50)
    writer.writerow([f_name,d_time,a_time,price])
f.close()
```

```python
# url 열기
url="https://www.yeogi.com/"
browser=webdriver.Chrome()# 새창열기
browser.maximize_window()# 창화면 확대
browser.get(url)
time.sleep(1)
```

```python
# # 2.스크롤내려 데이터 출력(이미지,숙소,평점,평가수,금액)
soup=BeautifulSoup(browser.page_source,'lxml')

hotels=soup.find_all('div',{'class':'gc-thumbnail-type-seller-card-wrapper css-1u8qly9'})
```

```python
# 2-6.스크롤내리기
prev_height=browser.execute_script("return document.body.scrollHeight")# 이전스크롤 높이
while True:# 현재스크롤 높이
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(3)# 추가데이터 로딩시간
    curr_height=browser.execute_script("return document.body.scrollHeight")
    if curr_height == prev_height:# 현재==이전까지 스크롤내림
        break
    # 같지 않으면
    prev_height=curr_height
    print('높이변화 :',curr_height)

print('스크롤 완료 :',curr_height)
```

```python
# 3.csv파일로 저장
title=['업소명','별점','리뷰','가격','썸네일']
f=open('hotelData1.csv','w',encoding='utf-8-sig',newline='')
writer=csv.writer(f)
writer.writerow(title)
f.close()
```

```python
# 2-7.숙박 리스트 출력
print('{}\t{}\t{}\t{}\t{}'.format(*title))
for hotel in hotels:
    try:
        thumbnail=hotel.find('img')['src']
        h_name=hotel.find('h3',{'class':'gc-thumbnail-type-seller-card-title css-1gxx2ac'}).text.strip()
        rate=hotel.find('span',{'class':'css-9ml4lz'}).text.strip()
        reviews=hotel.find('span',{'class':'css-9ml4lz'}).text.strip().split(' ')[0].strip().replace(",","")[:-1]
        price=hotel.find('span',{'class':'css-5r5920'}).text.strip().replace(",","")
        print(f'{h_name}\t{rate}\t{reviews}\t{price}\n{thumbnail}')
        print('-'*50)
        writer.writerow([h_name,rate,reviews,price,thumbnail])
        time.sleep(1)
    except:print('[오류]')
f.close()
```
